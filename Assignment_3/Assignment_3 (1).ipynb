## Graphs on few queries.

#### Comparision of reason of traveling through graph.


```python
Category =pd.crosstab(index=drive['CATEGORY*'],columns='Count of travels as per category')

Category.plot(kind='bar',color='b',alpha=0.8)
plt.legend()
Category
```

#### COMPARING THE NUMBER OF ROUND TRIPS vs SINGLE TRIPS


```python

plt.figure(figsize=(8,5))
def round(x):
    if x['START*'] == x['STOP*']:
        return 'Round Trip'
    else:
        return 'Single trip'
    
drive['KIND OF TRIP*'] = drive.apply(round, axis=1)

sns.countplot(drive['KIND OF TRIP*'], order=drive['KIND OF TRIP*'].value_counts().index, palette='viridis')
plt.show()
```

#### Ploting the count of purpose of booking the drive.


```python

plt.figure(figsize=(15,10))
sns.countplot(drive['PURPOSE*'],order=drive['PURPOSE*'].value_counts().index)
```

#### Graph for comparing the ride distance in business and personal category.


```python

drive.groupby('CATEGORY*')["MILES*"].mean().plot(kind='bar',color='y')
plt.axhline(drive["MILES*"].mean(),label='Mean distance travelled per ride')
plt.legend()
```

#### Graph to check trips across the month.

```python
plt.figure(figsize=(12,7))
sns.countplot(drive['MILES*'], order=drive['MILES*'].value_counts().index, palette='deep')
plt.axhline(drive['MILES*'].value_counts().mean(),linestyle='dotted', color = 'blue', label='Mean Trips across Months')
plt.legend()
plt.show()
```

#### Graph to check maximum stop location frequency for customers.

```python
plt.figure(figsize=(14,14))
pd.Series(drive['STOP*']).value_counts()[:25].plot(kind = "pie")
plt.title("Cab Rides STOP Location frequency")
plt.xticks(rotation = 50)
```

#### Graph to compare maximum number of rides in particular bucket of miles.

```python
function = {}

for p in drive['MILES*']:
    for p in drive['MILES*']:
        if p < 10:
            function.setdefault(p,'0-10 miles')
        elif p >= 10 and p < 20:
            function.setdefault(p,'10-20 miles')
        elif p >= 20 and p < 30:
            function.setdefault(p,'20-30 miles')
        elif p >= 30 and p < 40:
            function.setdefault(p,'30-40 miles')
        elif p >= 40 and p < 50:
            function.setdefault(p,'40-50 miles')
        else:
            function.setdefault(p,'Above 50 miles')
            
drive['BUCKET*'] = drive['MILES*'].map(function)

plt.figure(figsize=(10,6))
sns.countplot(drive['BUCKET*'], 
              order = drive['BUCKET*'].value_counts().index)
plt.show()
```
